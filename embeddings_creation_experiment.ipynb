```python
import colorsys
def generate_distinct_colors(num_colors):
    """
    Generate a list of distinct colors in HSL color space.
    
    Parameters:
    num_colors (int): The number of distinct colors to generate.
    
    Returns:
    list of str: A list of distinct colors in hexadecimal format.
    """
    colors = []
    for i in range(num_colors):
        hue = i / num_colors
        rgb = colorsys.hls_to_rgb(hue, 0.5, 1.0)  # Saturation and Lightness values can be adjusted
        r, g, b = [int(x * 255) for x in rgb]
        color_hex = "#{:02x}{:02x}{:02x}".format(r, g, b)
        colors.append(color_hex)
    return colors
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
class autoencoder(nn.Module):
    def __init__(self, input_dim):
        super(autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 8),
            nn.ReLU(),
            nn.Linear(8,4),
            nn.ReLU(),
            # nn.Linear(32,16),
            # nn.ReLU(),
            # nn.Linear(16,8),
            # nn.ReLU(),
            nn.Linear(4,2)
        )

        self.decoder = nn.Sequential(
            nn.Linear(2, 4),
            nn.ReLU(),
            # nn.Linear(8, 16),
            # nn.ReLU(),
            # nn.Linear(16, 32),
            # nn.ReLU(),
            nn.Linear(4, 8),
            nn.ReLU(),
            nn.Linear(8, input_dim),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        return self.decoder(x)
```

```python
class autoencoder1(nn.Module):
    def __init__(self, input_dim=82):
        super(autoencoder1, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            # nn.Linear(32, 16),
            # nn.ReLU(),
            # nn.Linear(16, 8),
            # nn.ReLU(),
            nn.Linear(32, 1),
        )
        self.decoder = nn.Sequential(
            nn.Linear(1, 32),
            nn.ReLU(),
            # nn.Linear(8, 16),
            # nn.ReLU(),
            # nn.Linear(16, 32),
            # nn.ReLU(),
            nn.Linear(32, input_dim),
            nn.Sigmoid()
        )
    def forward(self, x):
        x = self.encoder(x)
        return self.decoder(x)

class autoencoder2(nn.Module):
    def __init__(self, input_dim=13):
        super(autoencoder2, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 6),
            nn.ReLU(),
            nn.Linear(6, 2),
        )
        self.decoder = nn.Sequential(
            nn.Linear(2,6),
            nn.ReLU(),
            nn.Linear(6, input_dim),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        return self.decoder(x)

class EnsembleAutoEncoder(nn.Module):
    def __init__(self):
        super(EnsembleAutoEncoder, self).__init__()
        self.autoencoder1 = autoencoder1()
        self.autoencoder2 = autoencoder2()

    def forward(self, x):
        output1 = self.autoencoder1(x[:, :82])
        output2 = self.autoencoder2(x[:, 82:])
        ensemble_output = torch.cat((output1, output2), dim=1)
        return ensemble_output
```

```python
# Define a custom loss function with weights
class CustomLoss(nn.Module):
    def __init__(self, weights):
        super(CustomLoss, self).__init__()
        self.weights = weights

    def forward(self, y_true, y_pred):
        squared_errors = torch.square(y_true - y_pred)
        weighted_errors = self.weights * squared_errors
        return torch.mean(weighted_errors)
```

```python
def get_embeddings(data_loader, autoencoder):
    embeddings = []
    for batch in data_loader:
        with torch.no_grad():
            encoded = autoencoder.encoder(batch[0])
        embeddings.append(encoded)
    return torch.cat(embeddings)

def get_embeddings_from_ensemble(data_loader, ensemble_model):
    embeddings = []
    for batch in data_loader:
        with torch.no_grad():
            embeddings1 = ensemble_model.autoencoder1.encoder(batch[0][:, :82])
            embeddings2 = ensemble_model.autoencoder2.encoder(batch[0][:, 82:])
            
            # Combine the embeddings if needed
            combined_embeddings = torch.cat((embeddings1, embeddings2), dim=1)
        embeddings.append(combined_embeddings)
    return torch.cat(embeddings)
```
